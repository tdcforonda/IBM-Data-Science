### The function below takes a list containing either odd or even numbers, except one. The code then finds which number is the outlier. For example, we enter a list of odd numbers with one number being even, the code then returns the even number. I made this code in CodeWars as practice for Python.

````python
def find_outlier(integers):
    even = []
    even_numbers = 0
    odd = []
    odd_numbers = 0
    
    for number in integers:
        if number %2 == 0:
            even.append(number)
            even_numbers += 1
        elif number %2 == 1:
            odd.append(number)
            odd_numbers += 0
        
    if even_numbers == 1:
        return even[0]
    else:
        return odd[0]
    
print(find_outlier([160, 3, 1719, 19, 11, 13, -21]))
print(find_outlier([2, 4, 6, 8, 10, 3]))
````

**Alright, here we go**

1. First, we define the function using `def function_name()`
2. Inside the function, we define two lists and two counters. The counters will serve as the "finder" of the outlier. Basically, whichever is equal to one, then that's the outlier "type", which is either odd or even.
    
```
even = []
even_numbers = 0
odd = []
odd_numbers = 0
```

3. We then create a `for loop` that iterates through the list and checks if each number is divisible to two. If it is, then we put it in the list of even numbers, and if it is not, then we put it in the list of odd numbers. Every time that we add something to either list, the counter for their type goes up to "count" how many of their type exists.

```
for number in integers:
        if number %2 == 0:
            even.append(number)
            even_numbers += 1
        elif number %2 == 1:
            odd.append(number)
            odd_numbers += 0
```

4. Finally, we check which type has a count of exactly one number in the list. We then return the value of that outlier by accessing the first index of the list.

```
if even_numbers == 1:
        return even[0]
    else:
        return odd[0]
```
